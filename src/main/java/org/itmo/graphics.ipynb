```kotlin
%useLatestDescriptors
%use dataframe
%use kandy
```

```kotlin
val inputsSerialDf = DataFrame.readJson("../tmp/metrics/inputTimelineSerial.json")
val inputsParallelDf = DataFrame.readJson("../tmp/metrics/inputTimelineParallel.json")
val inputsDf = inputsParallelDf.concat(inputsSerialDf)
inputsDf
```

```kotlin
val resourcesDf = DataFrame.readJson("../tmp/metrics/resourcesTimeline.json")
resourcesDf
```

```kotlin
inputsDf.plot {
  line {
    x(v)
    y(time)
    color(isParallel)
  }

  layout.title = "execution time (ms) dependency from V (E = 3 * V)"
  layout.size = 1000 to 450
}
```

```kotlin
resourcesDf.plot {
  line {
    x(workers)
    y(time)
  }

  layout.title = "execution time (ms) dependency from workers count"
  layout.size = 1000 to 450
}
```
